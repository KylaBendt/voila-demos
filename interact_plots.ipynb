# Interactive Trig Plots

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
import numpy as np
from ipywidgets import interact
```

```python
def trigplot(frequency=1, amplitude=1, y_axis=1):
    """ Creates plots for tan(x), sin(x), and tan(x).
    y_axis determines the max and min y-scale.  
    This makes it easy to use the function with ipywidgets.interact since
    the scale does not change when you change the amplitude. 
    """
    f = frequency
    a = amplitude
    x = np.linspace(0, 2*np.pi, 1000)
    fig, subplts = plt.subplots(3,1, figsize=(10,10))
    # make a little extra space between the subplots
    fig.subplots_adjust(hspace=0.5)
    
    functions=(np.sin, np.cos, np.tan)
    titles = ('sin(x)', 'cos(x)', 'tan(x) = sin(x)/cos(x)')
    for ax, func, title in zip(subplts, functions, titles):
        ax.plot(x,a*func(x*f))
        ax.set_title(title)
        ax.axis([0, 2*np.pi, -y_axis,y_axis])
        major_ticks = np.arange(0,2*np.pi, .5*np.pi)
        ax.set_xticks(major_ticks)
        ax.grid(which='both');
```

```python
interact(trigplot, frequency=(1,10), amplitude= (1,3), y_axis=(1,10));
```
